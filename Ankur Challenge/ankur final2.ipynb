```python


def deleteFromStart(num, n): 

	# Get the number of digits 
	d = (math.log10(num) + 1); 

	# Declare a variable to form 
	# the reverse resultant number 
	rev_new_num = 0; 

	# Loop with the number 
	i = 0; 
	while (num != 0): 

		digit = num % 10; 
		num = int(num / 10); 

		if (i != (int(d) - n)): 
			rev_new_num = ((rev_new_num * 10) +
										digit); 
		i += 1; 

	# Declare a variable to form the 
	# resultant number 
	new_num = 0; 

	# Loop with the number 
	i = 0; 
	while (rev_new_num != 0): 

		new_num = ((new_num * 10) +
				(rev_new_num % 10)); 
		rev_new_num = int(rev_new_num / 10); 
		i += 1; 

	# Return the resultant number 
	return new_num; 

```

```python
def deleteFromEnd(num, n): 
  
    # Declare a variable to form  
    # the reverse resultant number 
    rev_new_num = 0; 
  
    # Loop with the number 
    i = 1; 
    while (num != 0): 
  
        digit = num % 10; 
        num = int(num / 10); 
  
        if (i != n): 
            rev_new_num = ((rev_new_num * 10) +
                                        digit); 
        i += 1; 
  
    # Declare a variable 
    # to form the resultant number 
    new_num = 0; 
  
    # Loop with the number 
    i = 0; 
    while (rev_new_num != 0): 
  
        new_num = ((new_num * 10) + 
                   (rev_new_num % 10)); 
        rev_new_num = int(rev_new_num / 10); 
        i += 1; 
  
    # Return the resultant number 
    return new_num; 
```

```python
def removing_unwanted(n):
    k = []
    for i in df["Telephone"]:
        try:
            i = i.replace("(", "")
            i = i.replace(")", "")
            i = i.replace(" ","")
            i = float(i)
            i = int(i)
            l = deleteFromStart(i,n)
            l = deleteFromStart(l,n)
            k.append(l)
        except:
            k.append("0")
        
    return k
```

```python
removed_unwanted_strings  = removing_unwanted(1)
```

```python
def greterthan_7(list,n):
    x = []
    for i in removed_unwanted_strings:
        i = str(i)
        while (len(i)>7):
            i = int(i)
            i = deleteFromEnd(i, n)
            i = str(i)
            if ((len(i))==7):
                break
        x.append(i)   
    return(x)
```
