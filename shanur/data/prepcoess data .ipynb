```python
test.columns = ["id","atribute"]
```

```python
test = test.drop(["id"],axis=1)
```

```python
data = []
for i in test["atribute"]:
    i = i.replace("    ", ",")
    i = i.replace("     ", ",")
    i = i.replace("  ", ",")
    i = i.replace(" ", ",")
    i = i = i.replace("''","")
    i = i = i.replace(",,",",")

    i = i.split(',')
    #print(i)
    data.append(i)
  
    
    
```
